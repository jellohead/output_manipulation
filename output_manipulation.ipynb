```python
import pandas as pd
import openpyxl

# Define the dataset
data = {
    'Column1': [1, 2, 3, 4, 5],
    'Column2': [6, 7, 8, 9, 10],
    'Column3': [11, 12, 13, 14, 15],
    'Column4': [16, 17, 18, 19, 20],
    'Column5': [21, 22, 23, 24, 25]
}

# Create a DataFrame from the dataset
df = pd.DataFrame(data)

# Create a new Excel workbook
excel_file = 'output.xlsx'
writer = pd.ExcelWriter(excel_file, engine='xlsxwriter')

# Select the desired sheet
sheet_name = 'Sheet1'
df.to_excel(writer, sheet_name=sheet_name, index=False)

# Load the workbook and select the sheet
book = openpyxl.load_workbook(excel_file)
sheet = book[sheet_name]

# Loop through the dataset columns and append as new columns to the sheet
start_column = df.shape[1] + 2  # Starting column for appending new columns

for column in df.columns:
    column_data = df[column]

    # Write the column data to the sheet as a new column
    for row, cell_value in enumerate(column_data, start=1):
        cell = sheet.cell(row=row, column=start_column)
        cell.value = cell_value

    start_column += 1  # Increment the starting column for the next loop

# Save the workbook
book.save(excel_file)
```
